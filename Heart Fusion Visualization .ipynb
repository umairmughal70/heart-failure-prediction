```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Dataset of Heart failure Prediction.csv') 
df=df.drop_duplicates()
df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
import seaborn as sns

sns.boxplot(df['RestingBP'])
plt.xlabel('Resting Blood Pressure')
plt.title('Box Plot of RestingBP')
plt.show()
```

```python
# Count heart disease df by sex
heart_disease_count = df[df['HeartDisease'] == 1]['Sex'].value_counts()

# Create bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=heart_disease_count.index, y=heart_disease_count.values)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Heart Disease Count by Sex')
plt.show()
```

```python
# Count heart disease df by sex
heart_disease_count = df[df['HeartDisease'] == 1]['ExerciseAngina'].value_counts()

# Create bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=heart_disease_count.index, y=heart_disease_count.values)
plt.xlabel('Exercise Angina')
plt.ylabel('Count')
plt.title('Heart Disease Count by Exercise Angina')
plt.show()
```

```python
# Count heart disease df by sex
heart_disease_count = df[df['HeartDisease'] == 1]['ST_Slope'].value_counts()

# Create bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=heart_disease_count.index, y=heart_disease_count.values)
plt.xlabel('ST_Slope')
plt.ylabel('Count')
plt.title('Heart Disease Count by ST_Slope')
plt.show()
```

```python
plt.pie(df['HeartDisease'].value_counts(), labels=df['HeartDisease'].unique(), autopct='%1.1f%%')
plt.title('Pie Chart of Heart Disease')
plt.show()
```

```python
plt.pie(df['ChestPainType'].value_counts(), labels=df['ChestPainType'].unique(), autopct='%1.1f%%')
plt.title('Pie Chart of ChestPainType')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Cholesterol'], bins=10)
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.title('Histogram of Cholesterol')
plt.show()
```

```python
plt.scatter(df['Age'], df['MaxHR'])
plt.xlabel('Age')
plt.ylabel('Resting Heart Rate')
plt.title('Scatter Plot of Age vs. MaxHR')
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=df['Sex'], y=df['Cholesterol'])
plt.xlabel('Sex')
plt.ylabel('Cholesterol')
plt.title('Box Plot of Sex vs. Cholesterol')
plt.show()
```

```python
import seaborn as sns

g = sns.FacetGrid(df, col='ChestPainType', col_wrap=2)
g.map(plt.hist, 'Cholesterol')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.suptitle('Histogram of Cholesterol for Each ChestPainType', y=0.95, fontsize=16, fontweight='bold')
plt.tight_layout(pad=2.5)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['Sex'] = labelencoder.fit_transform(df['Sex'])

labelencoder = LabelEncoder()
df['ChestPainType'] = labelencoder.fit_transform(df['ChestPainType'])

labelencoder = LabelEncoder()
df['RestingECG'] = labelencoder.fit_transform(df['RestingECG'])

labelencoder = LabelEncoder()
df['ST_Slope'] = labelencoder.fit_transform(df['ST_Slope'])

labelencoder = LabelEncoder()
df['ExerciseAngina'] = labelencoder.fit_transform(df['ExerciseAngina'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df.corr()

# Create a larger figure
plt.figure(figsize=(10, 8))

# Plot the correlation heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the data
data = {
    'Performance Matrices': ['Accuracy', 'Recall / Sensitivity', 'Specificity', 'Precision', 'ROC AUC'],
    'KNN [3]': [91.8, 88.6, 94.3, 92.1, 94.0],
    'Fusion Classifier': [95.3, 96.1, 99.4, 94.8, 96.0]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Set the 'Performance Matrices' column as the index
df.set_index('Performance Matrices', inplace=True)

# Plot the bar chart
ax = df.plot(kind='bar', rot=0)

# Set the title and labels
ax.set_title('Performance Matrices for Classification Models')
ax.set_xlabel('Performance Matrices')
ax.set_ylabel('Values')

# Add a legend
ax.legend(title='Classification Models')

# Show the plot
plt.show()
```
